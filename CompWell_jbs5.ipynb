```python
from datetime import date
from datetime import datetime
import datetime
import json
import os
import pandas as pd
import numpy as np
import glob
```

## Load jbs data from batch5

```python
all_jbs = glob.glob('/Users/giabraddock/compwell/batch5/Questionnaire/*/jbs_*')
jbs5 = pd.DataFrame()

# If file is empty, don't read
for f in all_jbs:
    data = pd.read_csv(f, header=0)
    if data.shape[0] == 0:
        continue
    else:
        jbs5 = pd.concat([jbs5, data], axis=0)

jbs5.sort_values('u_id', inplace=True)
jbs5.reset_index(drop=True,inplace=True)
```
